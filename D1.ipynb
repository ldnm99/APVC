```python
import numpy as np
import cv2
import os

#Loads only the images names in the current directory
def loader():
    name_img = []
    for filename in os.listdir():
        img = cv2.imread(os.path.join(filename))
        if img is not None:
            name_img.append(filename)
    return name_img

#Gives the name of the image file and returns classification 
#Image loading works
def classifier(name):
    img   = cv2.imread(name)
    hsv   = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    
    final = img.copy()
    
    sand_lower = np.array([10, 0, 0])
    sand_upper = np.array([30, 150, 255])
    
    sea_lower  = np.array([80, 0, 0])
    sea_upper  = np.array([120, 255, 255])
    
    sand_mask  = cv2.inRange(hsv, sand_lower, sand_upper)
    sea_mask   = cv2.inRange(hsv, sea_lower, sea_upper)
    
    mask   = cv2.bitwise_or(sand_mask, sea_mask)
    result = cv2.bitwise_and(final, final, mask=mask)
    
    nPix_sand = cv2.countNonZero(sand_mask)
    nPix_sea = cv2.countNonZero(sea_mask)
    print(nPix_sand,nPix_sea)
    
    #masks created
    cv2.imshow(name, result)
    
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    return "tudo puta"

#Runs the whole script
def main():
    names   = loader()
    results = np.column_stack((np.array(names), np.zeros(shape = (np.array(names).size))))
   
    for x in range(len(results)):
        output = classifier(results[x,0])
        results[x,1] = output
        
    print(results)
    
main()

```
